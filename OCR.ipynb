```python
import cv2
import matplotlib.pyplot as plt
import pytesseract

# Specify the path to the Tesseract executable
pytesseract.pytesseract.tesseract_cmd = '/opt/homebrew/bin/tesseract'

# Image processing
image_path = '/Users/shivam/Capstoone Project VITV/test1.jpg'
image = cv2.imread(image_path)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
threshold_value = 140
_, black_white = cv2.threshold(gray, threshold_value, 255, cv2.THRESH_BINARY)

# Display the processed image
plt.imshow(black_white, cmap='gray')
plt.axis('off')
plt.show()

# Save the processed image
output_path = '/Users/shivam/Capstoone Project VITV/processed_image.png'
cv2.imwrite(output_path, black_white)

# Perform OCR on the processed image
extracted_text = pytesseract.image_to_string(black_white)
print(extracted_text.strip())
```
